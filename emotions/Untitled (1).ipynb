```python
#WORKING but model not super accurate to emotions

# Load the trained model
#model = keras.models.load_model('path/to/trained/model.h5')
model = load_model('face_recognition/my_model_MobileNet.h5')

# Load the face and eye cascades
haar_cascade_face = cv2.CascadeClassifier('Face-Detection-in-Python-using-OpenCV-master/Face-Detection-in-Python-using-OpenCV-master/data/haarcascades/haarcascade_frontalface_default.xml')
haar_cascade_eye = cv2.CascadeClassifier('Face-Detection-in-Python-using-OpenCV-master/Face-Detection-in-Python-using-OpenCV-master/data/haarcascades/haarcascade_eye.xml')


# Open the webcam
cap = cv2.VideoCapture(0)

# Loop through frames from the webcam
while True:
    ret, frame = cap.read()


    # Detect faces in the frame
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    faces = haar_cascade_face.detectMultiScale(gray, 1.3, 5)

    # Loop through detected faces
    for (x, y, w, h) in faces:
        # Crop the face region
        face = frame[y:y+h, x:x+w]

        # Resize the face to the input size of the model
        face = cv2.resize(face, (224, 224))

        # Preprocess the face for the model
        face = cv2.cvtColor(face, cv2.COLOR_BGR2RGB)
        face = face.astype('float32') / 255.0
        face = np.expand_dims(face, axis=0)

        # Pass the face through the model to obtain the predicted emotion
        emotion_probabilities = model.predict(face)[0]
        emotion_label = np.argmax(emotion_probabilities)

        # Display the predicted emotion label on the frame
        emotion_text = ['Angry', 'Disgust', 'Fear', 'Happy', 'Sad', 'Surprise', 'Neutral'][emotion_label]
        cv2.putText(frame, emotion_text, (x, y-10), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)

        # Draw a rectangle around the face
        cv2.rectangle(frame, (x, y), (x+w, y+h), (255, 0, 0), 2)

    # Display the frame
    cv2.imshow('frame', frame)

    # Exit if 'q' is pressed
    if cv2.waitKey(1) == ord('q'):
        break

# Release the webcam and close all windows
cap.release()
cv2.destroyAllWindows()
```
